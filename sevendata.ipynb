## Convert timestamp to datetime 
Convert the Timestamp column to datatime format

```python
sheet2.dtypes
```

```python
sheet2['date_time'] = sheet2['timestamp'].astype('datetime64[ns]')
```

```python
sheet2.dtypes
```

### Get hour detail from time data

```python
sheet2['day'] = sheet2['date_time'].dt.day
sheet2['hour'] = sheet2['date_time'].dt.hour
```

```python
sheet2.head()
```

### converting date to dayofweek

```python
def day (x):
    if x == 20:
        return 'Monday'
    if x == 21:
        return 'Tuesday'
    if x == 22:
        return 'Wednesday'
    if x == 23:
        return 'Thursday'
    if x == 24:
        return 'Friday'
    if x == 25:
        return 'Saturday'
    if x == 26:
        return 'Sunday'
```

```python
sheet2['trans_day3'] = sheet2['day'].apply(day)
sheet2.head()
```

### converting hour to period of time (morning, afternoon, evening)

```python
def hour (x):
    if x == 10:
        return 'Morning'
    if x == 11:
        return 'Morning'
    if x == 13:
        return 'Afternoon'
    if x == 14:
        return 'Afternoon'
    if x == 17:
        return 'Evening'
    if x == 18:
        return 'Evening'
    if x == 19:
        return 'Evening'
```

```python
sheet2['hour2'] = sheet2['hour'].apply(hour)
sheet2.head()
```
